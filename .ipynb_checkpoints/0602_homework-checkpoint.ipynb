```python
!pip install selenium
!pip install chromedriver_autoinstaller
import chromedriver_autoinstaller
import time
import os
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from urllib.request import Request, urlopen
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)
url = 'https://banapresso.com/store'
driver.get(url)
time.sleep(3)
soup = BeautifulSoup(driver.page_source)
```

```python
store_list = []

# 1페이지부터 11페이지까지 반복
for page in range(0, 6):
        # 페이지 소스 가져오기
    page_source = driver.page_source
    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(page_source, 'html.parser')
        # 데이터 크롤링
    store_name_map = soup.findAll(['i', 'span'], {'class': 'store_name_map'})
    for i in store_name_map:
        name = i.find_next('i').text
        address = i.find_next('span').text
        store_list.append((name, address))  # 리스트에 추가
    # 페이지네이션 클릭
    pagination_buttons = driver.find_elements(By.CSS_SELECTOR, '.pagination > ul > li > a')
    if page < len(pagination_buttons):
        pagination_button = pagination_buttons[page]
        pagination_button.click()
    else:
        next_button = driver.find_element(By.CSS_SELECTOR, '.pagination > .btn_page_next > a')
        next_button.click()



# 전체 리스트 출력
for store in store_list:
    name, address = store
    print(name)
    print(address)
    print() 
```
